# 1장 

신뢰성 : 하드웨어, 소프트웨어, 인적오류에도 시스템이 원하는 성능 수준에서 기능수행

확장성 : 시스템에서 데이터의 트래픽, 복잡도가 증가해도 처리할 수 있음

유지보수성 : 다른 사람이 새로운 작업을 추가하거나 현재작업을 유지보수 가능해야함 

## 신뢰성 

결함 : 사양에서 벗어난 시스템의 한 구성요소

장애 : 필요한 서비스를 제공하지 못하고 멈춤
    
-> 결함으로 장애가 일어나지 않게끔 내결함성으로 설계

결함을 예측,대처할 수 있는 시스템 

-> 내결함성 or 탄력성 (내결함성은 모든 종류의 결함에 대처가능함을 의미(실제로 구현은 불가능))

### 하드웨어 오류

하드디스크의 평균 장애 시간 (MTTF = 가동시간/장애횟수) = 약 10~50년

<시스템 장애율을 줄이기 위한 대응>

1. 각 하드웨어의 구성요소에 중복을 추가함
-> 고가용성이 필수적인 소수의 애플리케이션에 사용

2. 소프트웨어 내결함성 기술 사용 및 하드웨어 중복성 추가 (aws와 같은 단일 장비 신뢰성보다 유연성, 탄력성을 우선시한 플랫폼)
-> 최근 추세, 장비 재부팅시 장비장애를 견딜 수 있으므로 전체 시스템의 중단시간 없이 한번에 한 노드씩 패치가능


### 소프트웨어 오류

소프트웨어의 체계적 오류 : 평상시에는 드러나지 않다가 특정 상황이 발생하면 드러남

   ex) cpu시간, 메모리 등 처럼 공유자원을 과도하게 사용하는 일부 프로세스
       한 구성요소의 결함이 다른 구성요소의 결함을 야기해 연쇄적으로 결함이 발생하는 연쇄과정
        
-> 신속한 해결책이 없다. 수행중에 지속적으로 확인해 경고를 발생시켜야 한다
   시스템의 가정과 상호작용에 대하여 생각, 빈트없는 테스트, 프로세스 격리, 죽은 프로세스의 재시작 허용, 
    프로덕션 환경에서 시스템 동작의 측정, 모니터링이 문제해결에 도움
   
    

### 인적 오류 

운영자의  설정 오류 

-> 오류 가능성 최소화한 설계 (잘 설계한 추상화, API, 관리 인터페이스)

   실제 데이터를 실험할 수 있지만 실제 사용자에게는 영향이 없는 '샌드박스'를 제공
   
   성능지표 마련

# 확장성 

시스템이 특정방식으로 커지면 어떻게 대처할까? 
추가 부하를 다루기위해 계산자원을 어떻게 투입할까?

### 부하 기술

부하 매개변수(시스템 설계에 따라 선정)로 나타냄


### 성능 기술

부하 매개변수가 증가했을때 성능에 어떻게 영향을 미칠지, 성능유지시 자원을 얼마나 늘려야할지 

-> 성능 수치기 필요

   하둡(일괄 처리 시스템)에서는 처리량에 관심
   
   온라인 시스템에서는 응답시간에 관심(같은 요청에 다양한 응답시간이 있으므로 분포로 생각해야함)
   
   꼬리 지연 시간(tail latency, 99분위같은 상위  백분위 응답시간)는 헤비유저가 경험하기 쉬우므로 중요한 지표이나 비용이 많이 듬
   
   높은 백분위에서 큐 대기 지연이 대부분이고 후속처리 요청을 처리해도 이전 요청이 완료되길 기다리는 시간인 
   선두 차단(head of line blocking) 발생

### 부하대응 접근방식

용량확장

수직확장

규모확장

수평확장

비공유 아키텍쳐


# 유지보수성

소프트웨어의 비용은 대부분 유지보수에서 기인

레거시 소프트웨어를 만들지 않게끔 설계원칙 설정

### 운용성

-> 운영팀이 시스템을 원활하게 운영할 수 있도록

### 단순성

-> 복잡도를 최대한 낮춤

### 발전성

-> 이후에 시스템 변경이 용이하도록

# 총정리

##기능성 요구 -> 원하는 기능 수행

## 비기능적 요구사항 -> 신뢰성, 확장성, 유지보수성

신뢰성 : 결함이 발생해도 시스템이 작동
    
확장성 : 부하가 증가해도 좋은 성능을 유지 
    
유지보수성 : 시스템의 상태를 잘 관찰가능, 효율적으로 관리
